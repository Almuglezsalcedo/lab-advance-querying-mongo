```python
import json
import pymongo
MongoClient = pymongo.MongoClient
client = MongoClient()
db = client.companies
```

```python
cursor=db.companies.find({'name':'Babelgum'})
list(map(lambda r: r["name"],list(cursor)))
```

```python
cursor2=db.companies.find({'number_of_employees':{'$gt':5000}}).sort("number_of_employees",pymongo.ASCENDING).limit(20)
list(cursor2)
```

```python
cursor3=db.companies.find({'founded_year':{'$in':[2000,2001,2002,2003,2004,2005]}})
list(map(lambda r: (r['name'],r['founded_year']),list(cursor3)))
```

```python
cursor4=db.companies.find({'$and': [{'ipo.valuation_amount':{'$gt':100000000}}, {'founded_year':{'$lt':2010}}]})
list(map(lambda r: (r['ipo']['valuation_amount'],r['name']),list(cursor4)))
```

```python
cursor5=db.companies.find({'$and': [{'number_of_employees':{'$lt':1000}}, {'founded_year':{'$lt':2005}}]}).sort("number_of_employees",pymongo.DESCENDING).limit(10)
list(cursor5)
```
